```python
import numpy as np

def SavePtsfile(filename,pts,head) :
   f=open(filename,"wt")
   print ('%d %d %d %d'%(head[0],head[1],head[2],head[3]),file=f)
   for i in range(len(pts)) :
       print ('%d %d %.3f %.3f %.3f'%(pts[i,0],pts[i,1],pts[i,2],pts[i,3],pts[i,4]),file=f)
   f.close()

Grille_0=np.array((np.loadtxt('Step11_Grille_0.txt')),dtype='float')
Grille_1=np.array((np.loadtxt('Step11_Grille_1.txt')),dtype='float')
Grille_2=np.array((np.loadtxt('Step11_Grille_2.txt')),dtype='float')
Grille_3=np.array((np.loadtxt('Step11_Grille_3.txt')),dtype='float')

Grille_1D=Grille_1
Grille_2D=Grille_2
Grille_3D=Grille_3

# ouverture fichier Translation genere a partir de la correlation entre 2-3: 
Translate=np.array((np.loadtxt('TranslationRef23_Step11.txt')),dtype='float')

Gradient=np.mat(Translate[0:3])
GradientReverse=np.mat(Translate[3:6])
AverageRefPos=np.mat(Translate[6]).reshape(3,1)
AverageCurrentPos=np.mat(Translate[7]).reshape(3,1)

# Changement de repere des points (DOWN vers UP)

Grille_3D[:,2:5]=np.transpose(((GradientReverse+np.eye(3,3)))*(np.transpose(Grille_3[:,2:5])-AverageCurrentPos)+AverageRefPos)

# ouverture fichier Translation genere a partir de la correlation entre 1-2: 
Translate=np.array((np.loadtxt('TranslationRef12_Step11.txt')),dtype='float')

Gradient=np.mat(Translate[0:3])
GradientReverse=np.mat(Translate[3:6])
AverageRefPos=np.mat(Translate[6]).reshape(3,1)
AverageCurrentPos=np.mat(Translate[7]).reshape(3,1)

# Changement de repere des points (DOWN vers UP)

Grille_3D[:,2:5]=np.transpose(((GradientReverse+np.eye(3,3)))*(np.transpose(Grille_3D[:,2:5])-AverageCurrentPos)+AverageRefPos)

Grille_2D[:,2:5]=np.transpose(((GradientReverse+np.eye(3,3)))*(np.transpose(Grille_2[:,2:5])-AverageCurrentPos)+AverageRefPos)

# ouverture fichier Translation genere a partir de la correlation entre 1-2: 
Translate=np.array((np.loadtxt('TranslationRef01_Step11.txt')),dtype='float')

Gradient=np.mat(Translate[0:3])
GradientReverse=np.mat(Translate[3:6])
AverageRefPos=np.mat(Translate[6]).reshape(3,1)
AverageCurrentPos=np.mat(Translate[7]).reshape(3,1)

# Changement de repere des points (DOWN vers UP)

Grille_3D[:,2:5]=np.transpose(((GradientReverse+np.eye(3,3)))*(np.transpose(Grille_3D[:,2:5])-AverageCurrentPos)+AverageRefPos)

Grille_2D[:,2:5]=np.transpose(((GradientReverse+np.eye(3,3)))*(np.transpose(Grille_2D[:,2:5])-AverageCurrentPos)+AverageRefPos)

Grille_1D[:,2:5]=np.transpose(((GradientReverse+np.eye(3,3)))*(np.transpose(Grille_1[:,2:5])-AverageCurrentPos)+AverageRefPos)
```

```python
# Parametres de la Grille 
Nbcolonnes=32
Nblignes=32
NbPlans=64

Grille_Total = np.zeros((Nbcolonnes*Nblignes*NbPlans,5))

Grille_Total[:18*Nbcolonnes*Nblignes] = Grille_0[:18*Nbcolonnes*Nblignes]

Grille_Total[18*Nbcolonnes*Nblignes:(18+15)*Nbcolonnes*Nblignes,0] = Grille_1D[Nbcolonnes*Nblignes:16*Nbcolonnes*Nblignes,0] + 17*Nbcolonnes*Nblignes
Grille_Total[18*Nbcolonnes*Nblignes:(18+15)*Nbcolonnes*Nblignes,1:] = Grille_1D[Nbcolonnes*Nblignes:16*Nbcolonnes*Nblignes,1:]

Grille_Total[(18+15)*Nbcolonnes*Nblignes:(18+15*2)*Nbcolonnes*Nblignes,0] = Grille_2D[Nbcolonnes*Nblignes:16*Nbcolonnes*Nblignes,0] + 32*Nbcolonnes*Nblignes
Grille_Total[(18+15)*Nbcolonnes*Nblignes:(18+15*2)*Nbcolonnes*Nblignes,1:] = Grille_2D[Nbcolonnes*Nblignes:16*Nbcolonnes*Nblignes,1:]

Grille_Total[(18+15*2)*Nbcolonnes*Nblignes:,0] = Grille_3D[Nbcolonnes*Nblignes:17*Nbcolonnes*Nblignes,0] + 47*Nbcolonnes*Nblignes
Grille_Total[(18+15*2)*Nbcolonnes*Nblignes:,1:] = Grille_3D[Nbcolonnes*Nblignes:17*Nbcolonnes*Nblignes,1:]

HeadOfFile=list(np.reshape(np.asarray([len(Grille_Total),NbPlans,Nbcolonnes,Nblignes]), (1,np.size([len(Grille_Total),NbPlans,Nbcolonnes,Nblignes])))[0])
SavePtsfile('Step11_Grille_Total.pts',Grille_Total,HeadOfFile)
```
