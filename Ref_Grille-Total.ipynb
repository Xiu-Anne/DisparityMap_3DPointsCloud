```python
CoordPoints_0 = np.array([[170.], [-170.], [-1498.]])
CoordPoints_1 = np.array([[2649.], [-170.], [-1498.]])
CoordPoints_2 = np.array([[170.], [-2649.], [-1498.]])
CoordPoints_3 = np.array([[2649.], [-2649.], [-1498.]])
CoordPoints_4 = PointE_UP
CoordPoints_5 = PointF_UP
CoordPoints_6 = PointG_UP
CoordPoints_7 = PointH_UP

def SavePts(filename,pts) :
   f=open(filename,"wt")
   for i in range(len(pts)) :
       print ('%d %d %.3f %.3f %.3f'%(pts[i,0],pts[i,1],pts[i,2],pts[i,3],pts[i,4]),file=f)
   f.close()

def SavePtsfile(filename,pts,head) :
   f=open(filename,"wt")
   print ('%d %d %d %d'%(head[0],head[1],head[2],head[3]),file=f)
   for i in range(len(pts)) :
       print ('%d %d %.3f %.3f %.3f'%(pts[i,0],pts[i,1],pts[i,2],pts[i,3],pts[i,4]),file=f)
   f.close()

# Parametres de la Grille 
Nbcolonnes=32
Nblignes=32
NbPlans=64

# Generation des points selon vecteurs AC, BD, EG et FH 

VecteursLignes=np.reshape(([CoordPoints_2-CoordPoints_0, CoordPoints_3-CoordPoints_1, CoordPoints_6-CoordPoints_4,CoordPoints_7-CoordPoints_5]),(4,3))

OrigineVecteursLignes=np.reshape(([CoordPoints_0,CoordPoints_1,CoordPoints_4,CoordPoints_5]),(4,3))

PointsVecteurs=np.ones((Nblignes,4*3),float)

for i in range(0,Nblignes) :
  StepL=(i/(Nblignes-1))
  for j in range(0,4) :
    PointsVecteurs[i,j*3:(j*3)+3]=(VecteursLignes[j]*StepL)+ OrigineVecteursLignes[j]

	
# Generation des points de la grille correspondant aux surfaces Sup et inf du parallelogramme a partir des pts obtenues selon les vecteurs precedents

SurfaceUP=np.ones((Nblignes*Nbcolonnes,3),float)
SurfaceDOWN=np.ones((Nblignes*Nbcolonnes,3),float)

for i in range(0,Nblignes) :
    for j in range(0,Nbcolonnes) :
      StepC=(j/(Nbcolonnes-1))
      SurfaceUP[(i*Nblignes)+j,0:3]=((PointsVecteurs[i,3:6]-PointsVecteurs[i,0:3])*StepC)+PointsVecteurs[i,0:3]
      SurfaceDOWN[(i*Nblignes)+j,0:3]=((PointsVecteurs[i,9:12]-PointsVecteurs[i,6:9])*StepC)+PointsVecteurs[i,6:9]

# Generation de tous les points de la grille a partir des points des surfaces Sup et Inf 


VolumeUpDown=np.zeros((Nblignes*Nbcolonnes*NbPlans,5),float)

for i in range(0,NbPlans) :
    StepP=(i/(NbPlans-1))
    VolumeUpDown[i*Nblignes*Nbcolonnes:(i+1)*Nblignes*Nbcolonnes,2:5]=((SurfaceDOWN-SurfaceUP)*StepP)+SurfaceUP
    VolumeUpDown[i*Nblignes*Nbcolonnes:(i+1)*Nblignes*Nbcolonnes,0]=np.arange(Nblignes*Nbcolonnes)+(i*Nblignes*Nbcolonnes)+1


# Enregistrement fichier point grille Totale dans le referenticiel courant (UP)

HeadOfFile=list(np.reshape(np.asarray([len(VolumeUpDown),NbPlans,Nbcolonnes,Nblignes]), (1,np.size([len(VolumeUpDown),NbPlans,Nbcolonnes,Nblignes])))[0])

SavePtsfile('testgrilleTotaleUP.pts',VolumeUpDown,HeadOfFile)
```

```python
# Enregistrement fichier point grille UP dans le referenticiel UP : Nbcolonnes*Nblignes x (19+16+15+14)
np0 = 19
np1 = 15 #+2
np2 = 15 #+2
np3 = 15 #+2

VolumeUP=VolumeUpDown[0:Nbcolonnes*Nblignes*np0,:]
VolumeDOWN1=VolumeUpDown[Nbcolonnes*Nblignes*(np0-2):Nbcolonnes*Nblignes*(np0+np1),:]
VolumeDOWN2=VolumeUpDown[Nbcolonnes*Nblignes*(np0+np1-2):Nbcolonnes*Nblignes*(np0+np1+np2),:]
VolumeDOWN3=VolumeUpDown[Nbcolonnes*Nblignes*(np0+np1+np2-2):,:]

# Enregistrement fichier point grille 0

HeadOfFile=list(np.reshape(np.asarray([len(VolumeUP),np0,Nbcolonnes,Nblignes]), ( 1,np.size([len(VolumeUP),np0,Nbcolonnes,Nblignes])))[0])
SavePtsfile('testgrille_0.pts',VolumeUP,HeadOfFile)

# Enregistrement fichier point grille 1-3

SavePts('testgrille_1.txt',VolumeDOWN1)
SavePts('testgrille_2.txt',VolumeDOWN2)
SavePts('testgrille_3.txt',VolumeDOWN3)
```
